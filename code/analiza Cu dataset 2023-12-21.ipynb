```python
colours = ['k','b','r','g', 'c', 'm','y', 'tab:purple', 'tab:pink','tab:brown', 'tab:orange', 'tab:gray', 'tab:cyan', 'xkcd:tan', 'xkcd:lilac']

plt.figure(figsize=(8,8))
ax = plt.axes(projection ="3d")
for d in range(len(drukarka)):
  ax.scatter3D(df[df['spektrometr']==drukarka[d]]['violet'],df[df['spektrometr']==drukarka[d]]['green'],df[df['spektrometr']==drukarka[d]]['yellow'],color = colours[d],label="level {}".format(drukarka[d]))
ax.set_xlabel('violet', color = 'm')
ax.set_ylabel('green', color = 'g')
ax.set_zlabel('yellow', color = 'y')
ax.view_init(10,20)
ax.legend(loc="upper right")
plt.title("Czujnik AS7262 - wybrane kolory (violet, green,yellow) 3D")
plt.show()
```

```python
#standaryzujemy
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
columns_to_normalize = ['violet','green','yellow']
columns_after_normalize = ['norm_violet','norm_green','norm_yellow']

x = df[columns_to_normalize].values
x_scaled = scaler.fit_transform(x)

df_temp = pd.DataFrame(x_scaled, columns = columns_after_normalize)
df[columns_after_normalize] = df_temp

df['zip'] = list(zip(df['norm_violet'],df['norm_green'],df['norm_yellow']))
df.head()
```

```python
df['zip1'] = list(zip(df['violet'],df['green'],df['yellow']))
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(len(drukarka))
```

```python
df['Labels'] = model.fit_predict(list(df['zip1']))
labels = np.unique(df['Labels'])
maximum = labels.max()
minimum = labels.min()
print('minimum: '+ str(minimum) +'    maximum: '+str(maximum))
print(df.shape)
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5))
fig.suptitle("Czujnik AS7262 - wybrane kolory (violet, green,yellow) 3D")
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.set_title("dane sklasteryzowane")
for d in range(maximum+1):
  ax.scatter3D(df[df['Labels']==d]['violet'],df[df['Labels']==d]['green'],df[df['Labels']==d]['yellow'],color = colours[d],label="cluster {}".format(d))
ax.set_xlabel('violet', color = 'm')
ax.set_ylabel('green', color = 'g')
ax.set_zlabel('yellow', color = 'y')
ax.view_init(10,20)
#ax.legend(loc="upper right")
bx = plt.subplot(1,2,1, projection='3d')
bx.set_title("wejściowe")
for d in range(len(drukarka)):
  bx.scatter3D(df[df['spektrometr']==drukarka[d]]['violet'],df[df['spektrometr']==drukarka[d]]['green'],df[df['spektrometr']==drukarka[d]]['yellow'],color = colours[d],label="level {}".format(drukarka[d]))
bx.set_xlabel('violet', color = 'm')
bx.set_ylabel('green', color = 'g')
bx.set_zlabel('yellow', color = 'y')
bx.view_init(10,20)
plt.show()
```
